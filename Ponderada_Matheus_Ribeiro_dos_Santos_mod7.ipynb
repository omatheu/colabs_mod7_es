## A escolha das linhas com diferentes intenções

```python
lists_query = dataset[dataset['intent'] == 'lists_query']
calendar_query = dataset[dataset['intent'] == 'calendar_query']
news_query_intent = dataset[dataset['intent'] == 'news_query']
```

- Escolhi as intenções listadas acima, porque acredito que façam mais sentido dentro do contexto do projeto do módulo. O usuário final da aplicação que estamos desenvolvendo, apesar de não pesquisar em inglês, fará pesquisas ou solicitações semelhantes a essas.

```python
lists_query_top5 = lists_query.iloc[0:5]
calendar_query_top5 = calendar_query.iloc[0:5]
news_query_top5 = news_query_intent.iloc[0:5]
```

- A criação de um novo dataset com as frases juntas

```python
new_df = pd.concat([lists_query_top5, calendar_query_top5, news_query_top5])
```

```python
new_df
```

## Adicionando mais colunas com novos parâmetros analisados dentro do dataset

- Os novos parâmetros que decidi aplicar na minha análise das frases de cada uma das intenções é o sentimento e NER (Entidades Nomeadas).

```python
import spacy
from textblob import TextBlob
```

```python
nlp = spacy.load('en_core_web_sm')

def extract_entities(text):
    doc = nlp(text)
    return [(ent.text, ent.label_) for ent in doc.ents]

new_df['entities'] = new_df['utt'].apply(extract_entities)
```

```python
print(new_df[['utt', 'entities']])
```

```python
def analyze_sentiment(text):
    return TextBlob(text).sentiment.polarity

new_df['sentiment'] = new_df['utt'].apply(analyze_sentiment)
```

```python
print(new_df[['utt', 'sentiment']])
```

- O valor de polaridade retornado por TextBlob(text).sentiment.polarity varia entre -1.0 e 1.0:

  - -1.0 indica um sentimento extremamente negativo.
  - 0 indica um sentimento neutro.
  - 1.0 indica um sentimento extremamente positivo.

## Aplicação de Regex para criar 2 novas colunas

```python
new_df['events'] = new_df['utt'].str.extract(r"\b(meeting|party|show|event|game)\b")
```

```python
new_df['category'] = new_df['utt'].str.extract(r"\b(news|technology|football|current events)\b")
```
